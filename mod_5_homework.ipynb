```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Установим seed для воспроизводимости результатов
np.random.seed(42)
# Создаем датасет о путешествиях
n_samples = 30 # Уменьшили количество строк до 30
destination = np.random.choice(['Париж', 'Рим', 'Токио'], size=n_samples)
accommodation = np.random.choice(['Отель', 'Аппартаменты', 'Хостел'], size=n_samples)
season = np.random.choice(['Весна', 'Лето', 'Осень', 'Зима'], size=n_samples)
days = np.random.randint(3, 15, size=n_samples)
cost_per_day = np.random.normal(150, 40, size=n_samples).round(2)
satisfaction = np.random.randint(1, 11, size=n_samples)
start_date = pd.to_datetime(pd.date_range(start='2023-01-01', periods=n_samples, freq='D'))
# Создаем DataFrame
small_travel_data = pd.DataFrame({
    'destination': destination,
    'accommodation': accommodation,
    'season': season,
    'days': days,
    'cost_per_day': cost_per_day,
    'satisfaction': satisfaction,
    'start_date': start_date
})

display(small_travel_data)
```

```python
# гипотеза: Средняя стоимость за день (cost_per_day) в летний сезон не отличается от средней стоимости за день в другие сезоны.


# Визуализация:
plt.figure(figsize=(12, 7))
sns.boxplot(x='season', y='cost_per_day', data=small_travel_data, palette='viridis')
plt.title('Распределение стоимости за день в зависимости от сезона (n=30)', fontsize=16)
plt.xlabel('Сезон', fontsize=12)
plt.ylabel('Стоимость за день (в евро)', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Средняя стоимость за день не зависит от сезона.
```
